### Data Type Conversions

**Task**: Convert the `USD_Production_Budget`, `USD_Worldwide_Gross`, and `USD_Domestic_Gross` columns to a numeric format by removing `$` signs and `,`.

```
remove = ['$', ',']
col_name = ['USD_Production_Budget', 'USD_Worldwide_Gross', 'USD_Domestic_Gross']

for i in col_name:
  for j in remove:
    data[i] = data[i].astype(str).str.replace(j, "")
  data[i] = pd.to_numeric(data[i])
data.info()
data
```

**Task**: Convert the `Release_Date` column to a Pandas Datetime type.

```
data.Release_Date = pd.to_datetime(data.Release_Date)
data.info()
```

### Separate the "old" (before 1969) and "New" (1970s onwards) Films

**Tasks**: Create two new DataFrames: `old_films` and `new_films`

# Run Your Own Regression with scikit-learn

$$ REV \hat ENUE = \theta _0 + \theta _1 BUDGET$$

```
regression = LinearRegression()

X = pd.DataFrame(new_films, columns=['USD_Production_Budget'])
y = pd.DataFrame(new_films, columns=['USD_Worldwide_Gross'])
regression.fit(X,y)
print(regression.intercept_)
print(regression.coef_)
print(regression.score(X,y))
```

**Tasks**: Run a linear regression for the `old_films`. Calculate the intercept, slope and r-squared. How much of the variance in movie revenue does the linear model explain in this case?

```
X1 = pd.DataFrame(old_films['USD_Production_Budget'])
y1 = pd.DataFrame(old_films['USD_Worldwide_Gross'])
reg = LinearRegression()
reg.fit(X1,y1)
print(reg.intercept_)
print(reg.coef_)
reg_score = reg.score(X1,y1)
print(reg_score)

```
